```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv(r"C:\Users\MS_More\Desktop\Data Science\Materials\A9\book.csv")
```

```python
data.head()
```

```python
frequent_items = apriori(data, min_support=0.2, use_colnames=True,max_len=3)
frequent_items
```

```python
model = association_rules(frequent_items, metric='lift',min_threshold=0.5)
model
```

```python
model.sort_values('lift',ascending=False)
```

```python
plt.scatter(model['support'],model['confidence'])
plt.xlabel('support')
plt.ylabel('confidence') 
plt.show()
```
